```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
credit_score_df = pd.read_csv("credit_score_new.csv")
credit_score_df.head()
```

```python
credit_score_df = credit_score_df.drop(['Type_of_Loan','Unnamed: 0','Annual_Income','Customer_ID','ID','Name','Occupation','Amount_invested_monthly'],axis = 1)
credit_score_df.head()
```

```python
credit_score_df['Credit_Score'] = credit_score_df['Credit_Score'].str.replace('Good', '2', n=-1)
credit_score_df['Credit_Score'] = credit_score_df['Credit_Score'].str.replace('Standard', '1', n=-1)
credit_score_df['Credit_Score'] = credit_score_df['Credit_Score'].str.replace('Poor', '0', n=-1)
credit_score_df['Credit_Score'] = credit_score_df[['Credit_Score']].apply(pd.to_numeric)
credit_score_df.head()
```

```python
credit_score_df.dtypes
```

```python
credit_score_df.nunique()
```

```python
# Generate our categorical variable lists
credit_cat = credit_score_df.dtypes[credit_score_df.dtypes == "object"].index.tolist()
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(credit_score_df[credit_cat]))

# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names(credit_cat)
encode_df.head()
```

```python
# Merge one-hot encoded features and drop the originals
credit_score_df = credit_score_df.merge(encode_df,left_index=True, right_index=True)
credit_score_df = credit_score_df.drop(credit_cat,1)
credit_score_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = credit_score_df["Credit_Score"].values
X = credit_score_df.drop(["Credit_Score"],1).values


# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train[0])
hidden_nodes_layer1 = 100
hidden_nodes_layer2 = 80
hidden_nodes_layer3 = 60


nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="sigmoid")
)


# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="sigmoid"))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="sigmoid"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model

# Import checkpoint dependencies
import os
from tensorflow.keras.callbacks import ModelCheckpoint

# Define the checkpoint path and filenames
os.makedirs("checkpoints/",exist_ok=True)
checkpoint_path = "checkpoints/weights.{epoch:02d}.hdf5"

# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Create a callback that saves the model's weights every epoch
cp_callback = ModelCheckpoint(
    filepath=checkpoint_path,
    verbose=1,
    save_weights_only=True,
    period = 5)
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled,y_train,epochs=50,callbacks=[cp_callback])
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save("AlphabetSoupCharity.h5")
```
