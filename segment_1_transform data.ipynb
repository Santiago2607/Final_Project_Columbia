```python
import pandas as pd
from datetime import datetime
```

```python
# import csv file 
df_creditscore = pd.read_csv("credit score.csv")
df_creditscore.head(10)
```

```python
# drop columns 
df_creditscore = df_creditscore.drop(['SSN','Credit_Mix','Changed_Credit_Limit'], axis = 1)
df_creditscore.reset_index(drop=True, inplace = True)
```

```python
# fill the blank cell with names
customer_name = pd.DataFrame({
 "Customer_ID": df_creditscore["Customer_ID"],
 "Name_New": df_creditscore["Name"],
 })
customer_name = customer_name.dropna()
customer_name = customer_name.drop_duplicates()
customer_name.reset_index(drop=True, inplace = True)

df_credit_score = df_creditscore.merge(customer_name, on = "Customer_ID",how = 'left')
df_credit_score = df_credit_score.drop(['Name'], axis = 1)
df_credit_score.head(5)
```

```python
# transform data
df_credit_score['Age'] = df_credit_score['Age'].astype(str).str.replace("_","")
df_credit_score['Age'] = df_credit_score['Age'].astype(int)
```

```python
customer_age = pd.DataFrame({
 "Customer_ID": df_credit_score["Customer_ID"],
 "Age_New": df_credit_score["Age"],
 })

```

```python
# drop the inappropriate ages 
customer_age = customer_age.dropna()
customer_age = customer_age.drop_duplicates()
customer_age.reset_index(drop=True, inplace = True)
customer_age = customer_age.loc[(customer_age['Age_New']<100) & (customer_age['Age_New'])>0]
customer_age.reset_index(drop=True, inplace = True)

df_credit_score = df_credit_score.merge(customer_age, on = "Customer_ID",how = 'left')
df_credit_score = df_credit_score.drop(['Age'], axis = 1)
df_credit_score.head(10)
```

```python
# clean the occupation column
df_credit_score['Occupation'] = df_credit_score['Occupation'].astype(str).str.replace("_______","0")
customer_job = pd.DataFrame({
 "Customer_ID": df_credit_score["Customer_ID"],
 "Occupation_New": df_credit_score["Occupation"],
 })

customer_job = customer_job.dropna()
customer_job = customer_job.drop_duplicates()
customer_job = customer_job.loc[customer_job["Occupation_New"]!= "0"]
customer_job.reset_index(drop=True, inplace = True)

df_credit_score = df_credit_score.merge(customer_job, on = "Customer_ID",how = 'left')
df_credit_score = df_credit_score.drop(['Occupation'], axis = 1)
df_credit_score.head(10)
```

```python
df_credit_score.reset_index(drop=True, inplace = True)
```

```python
# change the month to numerical data
month = df_credit_score["Month"]

for i in range(len(month)):
    mon_num =datetime.strptime(month[i],'%B').month
    month[i]=mon_num

df_credit_score["Month"] =  month
df_credit_score['Month'] = df_credit_score['Month'].astype(int)
```

```python
# transform annual income column
df_credit_score['Annual_Income'] = df_credit_score['Annual_Income'].astype(str).str.replace("_","")
df_credit_score['Annual_Income'] = df_credit_score['Annual_Income'].astype(float)
```

```python
# transform data
df_credit_score['Num_of_Loan'] = df_credit_score['Num_of_Loan'].astype(str).str.replace("_","")
df_credit_score['Num_of_Loan'] = df_credit_score['Num_of_Loan'].astype(float)
```

```python
# put the blank cells to not specified 
df_credit_score = df_credit_score.fillna(value = {"Type_of_Loan": "Not Specified"})
```

```python
# transform data
df_credit_score['Num_of_Delayed_Payment'] = df_credit_score['Num_of_Delayed_Payment'].astype(str).str.replace("_","")
df_credit_score['Num_of_Delayed_Payment'] = df_credit_score['Num_of_Delayed_Payment'].astype(float)
```

```python
# transform data
df_credit_score['Outstanding_Debt'] = df_credit_score['Outstanding_Debt'].astype(str).str.replace("_","")
df_credit_score['Outstanding_Debt'] = df_credit_score['Outstanding_Debt'].astype(float)
```

```python
# transform data
df_credit_score['Monthly_Balance'] = df_credit_score['Monthly_Balance'].astype(str).str.replace("_","")
df_credit_score['Monthly_Balance'] = df_credit_score['Monthly_Balance'].astype(float)
```

```python
# transform data
df_credit_score['Credit_History_Age'] = df_credit_score['Credit_History_Age'].astype(str).str.replace(' Years and ','.')
df_credit_score['Credit_History_Age'] = df_credit_score['Credit_History_Age'].astype(str).str.replace('Months','')
```

```python
# change the type of data
df_credit_score['Credit_History_Age'] = df_credit_score['Credit_History_Age'].astype(float)
```

```python
# transform data
error = "!@9#%8"
df_credit_score['Payment_Behaviour'] = df_credit_score['Payment_Behaviour'].astype(str).str.replace(error,"")
```

```python
# transform data
df_credit_score['Amount_invested_monthly'] = df_credit_score['Amount_invested_monthly'].astype(str).str.replace('_',"")
df_credit_score['Amount_invested_monthly'] = df_credit_score['Amount_invested_monthly'].astype(float)
```

```python
# compute the isnull number 
df_credit_score.isnull().sum()
```

```python
# drop null values 
df_credit_score = df_credit_score.dropna()
df_credit_score.isnull().sum()
```

```python
# change column names and order 
df_credit_score = df_credit_score.rename(columns={'Name_New': 'Name',
                                                  'Age_New':'Age' ,
                                                  'Occupation_New':'Occupation'
                                                 })
cols_new = ['ID',
 'Customer_ID',
 'Name',
 'Age',
 'Occupation',    
 'Month',
 'Annual_Income',
 'Monthly_Inhand_Salary',
 'Num_Bank_Accounts',
 'Num_Credit_Card',
 'Interest_Rate',
 'Num_of_Loan',
 'Type_of_Loan',
 'Delay_from_due_date',
 'Num_of_Delayed_Payment',
 'Num_Credit_Inquiries',
 'Outstanding_Debt',
 'Credit_Utilization_Ratio',
 'Credit_History_Age',
 'Payment_of_Min_Amount',
 'Total_EMI_per_month',
 'Amount_invested_monthly',
 'Payment_Behaviour',
 'Monthly_Balance',
 'Credit_Score'
]
```

```python
# create customer table 
df_credit_score = df_credit_score[cols_new]
df_credit_score.reset_index(drop=True, inplace = True)
customer_cols = [
    'Customer_ID',
    'Name',
    'Age',
    'Occupation',
    ]
customer_info_df = df_credit_score[customer_cols]
customer_info_df = customer_info_df.drop_duplicates(subset=['Customer_ID'])
customer_info_df
```

```python
# create new credit change table 
loan_cols = [
  'ID',
 'Customer_ID', 
 'Month',
 'Annual_Income',
 'Monthly_Inhand_Salary',
 'Num_Bank_Accounts',
 'Num_Credit_Card',
 'Interest_Rate',
 'Num_of_Loan',
 'Type_of_Loan',
 'Delay_from_due_date',
 'Num_of_Delayed_Payment',
 'Num_Credit_Inquiries',
 'Outstanding_Debt',
 'Credit_Utilization_Ratio',
 'Credit_History_Age',
 'Payment_of_Min_Amount',
 'Total_EMI_per_month',
 'Amount_invested_monthly',
 'Payment_Behaviour',
 'Monthly_Balance',
 'Credit_Score'
    ]
credit_df = df_credit_score[loan_cols]
credit_df
```

```python
# optional
datasets_final = credit_df.merge(customer_info_df, on = "Customer_ID",how = 'left')
datasets_final.shape
```
